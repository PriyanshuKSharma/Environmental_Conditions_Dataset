```python
import pandas as pd
import numpy as np
```

```python
#Name of the cities in India
cities = [
    'Delhi', 'Mumbai', 'Kolkata', 'Chennai', 'Bangalore', 'Hyderabad', 'Ahmedabad',
    'Pune', 'Surat', 'Jaipur', 'LucknowUntitled document', 'Kanpur', 'Nagpur', 'Visakhapatnam',
    'Bhopal', 'Patna', 'Ludhiana', 'Agra', 'Nashik', 'Vadodara', 'Varanasi',
    'Srinagar', 'Aurangabad', 'Dhanbad', 'Amritsar', 'Navi Mumbai', 'Allahabad',
    'Howrah', 'Ranchi', 'Gwalior', 'Jabalpur', 'Coimbatore', 'Vijayawada', 
    'Jodhpur', 'Madurai', 'Raipur', 'Kota', 'Guwahati', 'Chandigarh', 'Solapur'
]
```

```python
# Number of data points per city
num_data_points = 100# Generate synthetic data
data = {
    'City': np.random.choice(cities, num_data_points * len(cities)),
    'Date': pd.date_range(start='1/1/2013', periods=num_data_points * len(cities), freq='D'),
    'PM2.5': np.random.uniform(10, 300, num_data_points * len(cities)),
    'PM10': np.random.uniform(20, 400, num_data_points * len(cities)),
    'NO2': np.random.uniform(10, 150, num_data_points * len(cities)),
    'SO2': np.random.uniform(5, 100, num_data_points * len(cities)),
    'CO': np.random.uniform(0.1, 10, num_data_points * len(cities)),
    'O3': np.random.uniform(10, 100, num_data_points * len(cities)),
    'Temperature': np.random.uniform(10, 45, num_data_points * len(cities)),
    'Humidity': np.random.uniform(20, 100, num_data_points * len(cities))
}
```

```python
# Create DataFrame
df = pd.DataFrame(data)
```
